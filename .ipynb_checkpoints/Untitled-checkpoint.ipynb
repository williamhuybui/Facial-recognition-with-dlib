```python
import io
import pandas as pd
from PIL import Image,ImageDraw, ImageFont
import face_recognition
import numpy as np
```

```python
def drawVertices(image_source, locations):
    pillow_img = Image.open(io.BytesIO(image_source))
    draw = ImageDraw.Draw(pillow_img)
    font = ImageFont.truetype('/Library/Fonts/Arial.ttf', 30) #Maybe different on non-Mac

    for i in locations: #top, left, bottom, right
        top, left, bottom, right=i
        p1=(top,right)
        p2=(top,left)
        p3=(bottom,left)
        p4=(bottom,right)
        for vertices in ((p1,p2),(p2,p3),(p3,p4),(p4,p1)):
            draw.line(((vertices[0][1], vertices[0][0]), (vertices[1][1], vertices[1][0])),
                        fill='green',
                        width=8
                )
#         draw.text((vertices[0].x + 10, vertices[0].y),
#               font=font, text=text_list[index], 
#               fill=(255, 255, 255))
    return pillow_img
```

```python
#All the image in train set
from os import listdir
from os.path import isfile, join

train_folder_dir="train"
jpg_list = [train_folder_dir+'/'+f for f in listdir(train_folder_dir) if (f.split(".")[1] in ["jpg","jpeg"])]
```

```python
2*np.ones(2)
```

```python
df_raw=pd.DataFrame({'array':[],'label':[]})
array_list=[]
for image_dir in jpg_list:
    print(image_dir)
    image=face_recognition.load_image_file(image_dir)
    face_locations = face_recognition.face_locations(image)

    for face_location in face_locations:
        
        top, right, bottom, left = face_location
        face_image = image[top:bottom, left:right]
        array_list.append(face_image)
#         pil_image = Image.fromarray(face_image)
#         pil_image.save(str(count)+'.jpg')

#     pil_image = Image.fromarray(face_image)
#     pil_image.show()
df_raw['array']=array_list
df_raw['label']=np.ones(len(array_list))*-1
```
